```python
import yfinance as yf
import pandas as pd
import datetime
from openpyxl import *

# Define the list of ticker symbols
tickers = ['JPM', 'GS', 'MS']

# Define the start and end dates
start_date = '2017-01-01'
end_date = '2022-03-31'

# Get the current date
current_date = datetime.datetime.now().strftime("%Y-%m-%d")

# Create a Pandas Excel writer
workbook = Workbook()
writer = pd.ExcelWriter(current_date + '_bank_stock_data' + '.xlsx', engine='openpyxl')
writer.book = workbook

# Iterate over each ticker symbol
for ticker in tickers:
    # Fetch the historical stock data using yfinance
    stock_data = yf.download(ticker, start=start_date, end=end_date)

    # Create a DataFrame from the stock data
    df = pd.DataFrame(stock_data)

    # Write the DataFrame to a new sheet in the Excel file
    df.to_excel(writer, sheet_name=ticker)

# Save the Excel file
file_path = current_date + '_bank_stock_data' + '.xlsx'
password = 'Team_17_SVB'

writer.save()
writer.close()

# Load the workbook with openpyxl
wb = load_workbook(file_path)

# Set the password to encrypt the workbook
wb.security.workbookPassword = password

# Save the encrypted workbook
wb.save(file_path)
print("Stock data saved and encrypted to " + file_path)
```
